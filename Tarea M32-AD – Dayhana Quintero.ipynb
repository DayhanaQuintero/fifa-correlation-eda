```python
#Importar un archivo csv con Pandas

#Se importa la libreria Pandas
import pandas as pd
#Otras librerias
import numpy as np
import os

df=pd.read_csv('fifa_eda.csv')
```

```python
corr_fifa=df.select_dtypes(include="number").corr()
corr_fifa
```

```python
import plotly.express as px
fig = px.imshow(corr_fifa, text_auto=True,title="Heatmap de correlación – FIFA",  width=1000,height=1000)

fig.show()
```

# Elemento 2: Gráfico de Correlación que responda – “Cuál es la relación entre la edad y el overall?” – explicar la salida 

```python
fig = px.scatter(df, x="Age", y="Overall", color="Overall", color_continuous_scale="Viridis", trendline="ols", title="Relación entre Edad y Overall")
fig.show()
```

- Este gráfico muestra la relación entre la edad de los jugadores y su calificación general (Overall).
- Se observa que el Overall tiende a aumentar en edades cortas, alcanzando su punto máximo aproximadamente entre los 26 y 30 años, y posteriormente comienza a disminuir conforme aumenta la edad.
- La línea de tendencia indica una relación donde los jugadores jóvenes y muy mayores presentan, en promedio, calificaciones más bajas que los de edades intermedias.
- En general el rendimiento está asociado a la edad, con un mejor desempeño en la etapa de madurez deportiva.

# Elemento 3: Generar un gráfico de barras por club que indique el número de jugadores 

```python
df_club_count = df.groupby("Club").size().reset_index(name="Num_Jugadores")
```

```python
fig=px.bar(df_club_count, x="Club", y="Num_Jugadores", color="Club", title="Número de jugadores por club", width=1500,height=500)
fig.show()
```

# Elemento 4: Hacer un gráfico multipanel que indique la relación altura (height) vs skill moves, siendo la variable del panel si es zurdo o derecho 

```python
fig = px.strip(df,x="Height",y="Skill Moves",facet_col="Preferred Foot",color="Preferred Foot",
               title="Relación entre Altura y Skill Moves según pie dominante")
fig.show()
```

# Elemento 5: Un gráfico que crea importante para mostrar a los potenciales cracks

```python
cracks = df[df["Potential"] >= 85]

fig = px.scatter(cracks, x="Age",y="Potential",color="Club",hover_name="Name",title="Jugadores con mayor potencial (posibles cracks)")
fig.show()
```
